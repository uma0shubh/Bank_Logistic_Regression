## y -> Whether the client has subscribed a term deposit or not 
## Binomial ("yes" or "no")

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
# Import dataset
bank=pd.read_csv('bank-full.csv',sep=';')
bank.head()
```

### EDA

```python
bank.info()
```

```python
# One-Hot Encoding of categrical variables
data1 = pd.get_dummies(bank,columns=['job','marital','education','contact','month','poutcome'])
data1
```

```python
# To see all columns
pd.set_option("display.max.columns", None)
data1
```

```python
data1.info()
```

```python
# lABEL encoding - Custom Binary Encoding of Binary o/p variables 
data1['default'] = np.where(data1['default'].str.contains("yes"), 1, 0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"), 1, 0)
data1['y'] = np.where(data1['y'].str.contains("yes"), 1, 0)
data1
```

```python
data1.info()
```

### Model Building

```python
# Dividing our data into input and output variables
x = pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:]],axis=1)
y = data1.iloc[:,10]
```

```python
# Logistic regression model
classifier = LogisticRegression()
classifier.fit(x,y)
```

### Model Predictions

```python
# Predict for x dataset
y_pred = classifier.predict(x)
y_pred
```

```python
y_pred_df = pd.DataFrame({'actual_y':y,
                          'y_pred_prob':y_pred})
y_pred_df
```

### Testing Model Accuracy

```python
# Confusion Matrix for the model accuracy
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
# model accuracy = (a+d)/(a+b+c+d)
(39156+1162)/(39156+766+4127+1162)
```

```python
#Classification report
from sklearn.metrics import classification_report
print(classification_report(y,y_pred))
```

```python
# As accuracy = 0.8918, which is greater than 0.5 
# Thus, Threshold value>0.5=1 else Threshold value<0.5=0 
classifier.predict_proba(x)[:,1] 
```

```python
# ROC Curve plotting and finding AUC value
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
